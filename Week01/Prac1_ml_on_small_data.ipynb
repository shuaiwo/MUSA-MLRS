### Example: a more advanced outlier detection

```python
# Local Outlier Factor (LOF) method from the sklearn library to detect outliers in the 'tsRel' column of a pandas DataFrame dfTmp

from sklearn.neighbors import LocalOutlierFactor # imports the LocalOutlierFactor class from the sklearn library.

# fit the model for outlier detection (default)

# converts the 'tsRel' column of dfTmp into a 2D array
# reshapes the array into a 2D array with one column and as many rows as there are in dfTmp.
X = np.array(dfTmp.tsRel).reshape(dfTmp.shape[0],1)
X.shape

# n_neighbors parameter is the number of neighbors to use for calculating the local density and the local outlier factor
# contamination is the proportion of outliers in the data set
clf = LocalOutlierFactor(n_neighbors=5, contamination=0.1)
clf.fit_predict(X) # fits the model to the data X and predicts the labels (1 for inliers, -1 for outliers) of X according to the LOF algorithm.
dfTmp['outScore'] = clf.negative_outlier_factor_.tolist()
dfTmp
```

```python
dfTmp.plot.scatter(x='tsRel', y='outScore', c='DarkBlue')
```

### Correlations for filtered data

```python
# returns a Boolean Series where each value is True if the corresponding value in the 'outScore' column 
# of dfTmp is greater than -5 and less than 5, and False otherwise.

dfTmpFil = dfTmp[np.logical_and(dfTmp.outScore>-5, dfTmp.outScore<5)]
dfTmpFil
```

### Normalize data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#scaler = StandardScaler()
scaler = MinMaxScaler() # MinMaxScaler transforms features by scaling each feature to a given range, typically between zero and one.

scaler.fit(X) # computes the minimum and maximum to be used for later scaling
Xnorm = scaler.transform(X) # performs the scaling of the data
Xnorm.max(axis=0) # returns the maximum value of each column in the array Xnorm
```

#### Cross validation with leave one out

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut() # LeaveOneOut is a cross-validation method that splits the data into training and test sets by leaving out one observation at a time.
for train_index, test_index in loo.split(Xnorm): # loo.split(Xnorm) returns the indices of the training and test sets for each iteration of the cross-validation
    print('TRAIN: ' + str(train_index) + '  TEST: ' +  str(test_index))
```

```python
import warnings
warnings.filterwarnings('ignore') # suppresses warnings

from sklearn.svm import LinearSVR

loo = LeaveOneOut()

predAll = np.zeros([y.shape[0],1]) # creates an array of zeros with the same number of rows as y and one column

i=0
for train_index, test_index in loo.split(Xnorm):

    X_train, X_test = Xnorm[train_index], Xnorm[test_index] # selects the rows of Xnorm that correspond to the training and test sets
    y_train, y_test = y[train_index], y[test_index] # selects the rows of y that correspond to the training and test sets

    regr = LinearSVR(random_state=0, tol=1e-5) # creates a LinearSVR model
    
    regr.fit(X_train, y_train)  #  Train the model

    ypred = regr.predict(X_test)  #  Apply the model
    
    predAll[i] = ypred
    
    print('Y : ' + str(y_test) + ' , pred as: ' + str(ypred))
    i = i + 1 
```

```python
np.corrcoef(y.T, predAll.T) # computes the correlation coefficient between the actual and predicted values of y
# Pearson correlation coefficient is a measure of the linear correlation between two variables. It has a value between +1 and -1,
```
